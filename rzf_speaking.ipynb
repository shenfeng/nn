```python
import glob
txt = [open(l).read() for l in glob.glob('Ren-master/*/*.md')]
```

```python
from collections import Counter
chars = Counter()
for t in txt:
    chars.update(t)

```

```python
cc = chars.most_common()
```

```python
cc[:10]
```

```python
import sentencepiece as spm
import io

model = io.BytesIO()
spm.SentencePieceTrainer.Train(sentence_iterator=iter(txt), model_writer=model, vocab_size=3000, character_coverage=0.9999, byte_fallback=True)
```

```python
sp = spm.SentencePieceProcessor(model_proto=model.getvalue())
print(sp.get_piece_size())

for i in range(sp.get_piece_size()):
    if i > 255:
        if len(sp.IdToPiece(i)) > 1:
            print(i, sp.IdToPiece(i))
        
```
